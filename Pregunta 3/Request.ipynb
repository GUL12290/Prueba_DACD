# Pruebas

Con este Notebook, haremos pruebas de solicitudes a la API para verificar que todo funcione correctamente

## Prueba de Post con Json

```python
import requests
import json

url = 'http://127.0.0.1:5000/ejercicio/descarga/'

data = {
    "folio": "789",
    "nombre": "Chris",
    "ciudad_nacimiento": "Lisboa",
    "empresa": "Microsoft",
    "trabajo": "Ventas",
    "dirección": "Avenida de oro 67",
    "estado_destino": "Puebla",
    "cell": "555-198-4863",
    "fecha": "2023-06-19",
    "clasificacion": "A",
    "soltero": False,
    "formato": "json"
}

headers = {
    "Content-Type": "application/json"  # Set the Content-Type header to application/json
}

response = requests.post(url, data=json.dumps(data), headers=headers)

if response.status_code == 201:
    print("User added successfully.")
else:
    print("Error:", response.text)
```

## Prueba de Post con CSV

```python
import requests
import json

url = 'http://127.0.0.1:5000/ejercicio/descarga/'

data = {
    "folio": "789",
    "nombre": "Chris",
    "ciudad_nacimiento": "Lisboa",
    "empresa": "Microsoft",
    "trabajo": "Ventas",
    "dirección": "Avenida de oro 67",
    "estado_destino": "Puebla",
    "cell": "555-198-4863",
    "fecha": "2023-06-19",
    "clasificacion": "A",
    "soltero": False,
    "formato": "csv"
}

headers = {
    "Content-Type": "application/json"  # Set the Content-Type header to application/json
}

response = requests.post(url, data=json.dumps(data), headers=headers)

if response.status_code == 201:
    print("User added successfully.")
elif response.status_code == 200:
    print("User added successfully.")
else:
    print("Error:", response.text)
```

## Prueba de Get como csv

```python
import requests

url = 'http://127.0.0.1:5000/ejercicio/descarga/?formato=csv'
response = requests.get(url)

if response.status_code == 200:
    with open('datos.csv', 'wb') as file:
        file.write(response.content)
    print("Archivo CSV descargado exitosamente.")
else:
    print("Error al descargar el archivo CSV:", response.text)
```

## Prueba de Get como Json

```python
import requests

url = 'http://127.0.0.1:5000/ejercicio/descarga/?formato=json'  
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    # Process the JSON data as needed
    print("JSON data:", data)
else:
    print("Error:", response.text)
```
